## Question 5
The null hypothesis is that humans don't have a left or right head tilt tendency when kissing (just a random movement). Based on the code below, the p-value derived is most often than not, in between 0.1 and 0.05, which means that we have weak evidence against the null hypothesis and thus, showing that often couples have a right of left head tilt tendency but not always. 

```python
import random
from scipy.stats import chisquare

# Define the head tilting simulation for couples
def couple_head_tilt():
    # Each partner can tilt left (0) or right (1), 50/50 probability
    person1 = random.choice(["Right", "Left"])
    person2 = random.choice(["Right", "Left"])
    return (person1, person2)

# Set a random seed for reproducibility
random.seed(45322)

# Simulate head tilting for 124 couples
tilt_combinations = [couple_head_tilt() for _ in range(124)]

# Count occurrences of each combination
counts = {
    ("Right", "Right"): 0,
    ("Right", "Left"): 0,
    ("Left", "Right"): 0,
    ("Left", "Left"): 0
}

for couple in tilt_combinations:
    counts[couple] += 1

# Observed counts
observed_counts = list(counts.values())

# Expected counts assuming random 50/50 tilting for each partner
expected_prob = [0.25, 0.25, 0.25, 0.25]  # Each of the 4 outcomes is equally likely
expected_counts = [0.25 * 124] * 4  # Expected count for each outcome

# Perform chi-square test
chi2, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Output observed counts and p-value
observed_counts, p_value
```

## Question 8
**Problem Introduction**

I am tasked with finding out whether the 49 out of 80 STA130 students correctly guessed if I poured milk or tea first, or if they were randomly guessing and got it correct. 

**Relationship with Fisher's Tea**

This experiment is based off of the content's from Fisher and Bristol's experiment where Fisher had Bristol drink 8 cups of tea with 4 having milk poured first. She was able to correctly guess which cups had milk before tea, which is the basis of this experiment. The only difference is that the sample size and number of trials is larger, but the results should remain the same assuming Bristol correctly identified the cups based on taste.

**Null and Alternative Hypothesis**

The null hypothesis in this experiment is that the 49 students who correctly stated which drink was poured first were randomly guessing. That is, they did not know whether they were right or wrong and placed a random guess. The alternative hypothesis is that the 49 students who correctly identified the cups could actually taste the difference and made a decision based on the flavour. That means they were not randomly guessing and could identify the difference. So to reject the null hypothesis, I have to see a p-value not close to 0.5 since that would mean they are not making a random 50/50 guess, but actually an educated one.

Formally, the null hypothesis suggests that the p-value should be 0.5, that is, the population proportion of students who correctly guess made a 50/50 random guess. The alternative hypothesis refutes that by saying the p-value is not 0.5, which suggests the students are better or worse at guessing but shows that it wasn't a random guess.
- The population in this experiment is all STA130 students, but we are only sampling 80 of them to see if they will provide evidence against our null hypothesis
- Our parameter of interest is the population proportion, which represents all of the students who correctly guessed if milk or tea was poured first. To test this, we'll use the observed test statistic of the z-test for proportions to see how many standard deviations the 49 student's correct guess is away from the p-value

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 80  # number of students
p_null = 0.5  # probability under the null hypothesis
observed_p = 49 / 80  # observed proportion (49 correct out of 80)
num_simulations = 10000  # number of simulations
np.random.seed(42)  # for reproducibility

# Simulate sampling under the null hypothesis
simulated_proportions = np.random.binomial(n, p_null, num_simulations) / n

# Calculate p-value
extreme_proportions = np.abs(simulated_proportions - p_null) >= np.abs(observed_p - p_null)
p_value = np.mean(extreme_proportions)

# Output the result
print(f"Simulated p-value: {p_value}")

# Plotting the simulation results
plt.hist(simulated_proportions, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(observed_p, color='red', linestyle='dashed', linewidth=2, label=f'Observed p = {observed_p:.3f}')
plt.axvline(p_null, color='blue', linestyle='dashed', linewidth=2, label='Null p = 0.5')
plt.title('Sampling Distribution under Null Hypothesis')
plt.xlabel('Proportion of Correct Answers')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

In the code above, I simulated the experiment many times to see if the results would change. Referencing the graph, I can tell that the proportion of correct answers don't always fall in 0.5, but on either side of that value. That suggests that many students weren't flat out guessing and made a sort of educated guess since the p-value is not equal to 0.5. Additionally, the red dotted line being far from the center suggests that the results from this graph are unlikely to occur just by random guessing. So it's safe the reject the null hypothesis that the 49 students who correctly identified the cups were randomly guessing, but were actually determining by taste. Our observed test statistic ofo the z-test also gave us around 2.01, which represents the amount of standard deviations the observed sample proportion was from the hypothesized proportion of 0.5. Since the z-test of 2.01 is greater than 1.96 (the value for there to be a significant difference), we can reject the null hypothesis because there is numerical evidence that the observed proportion of students who correctly identified the cups is significantly different from the 0.5 guessing rate. Since our simulated p-value is a really small number, it shows that the probability of the null hypothesis happening is also really small.
